## 本文件用于可视化系统矩阵数据

```python
import numpy as np
import matplotlib.pyplot as plt

with open('PSF2D_100x100x100x120.dat', 'rb') as f:
    data = np.array(np.fromfile(f, dtype=np.float32))
print (data)
print(data.shape)
data=data.reshape(1,120,100,100,100)
print(data.shape)


fig, ax = plt.subplots()
cax = ax.imshow(data[0,45,50], interpolation='nearest', cmap='viridis')
fig.colorbar(cax)
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
plt.show()
```
